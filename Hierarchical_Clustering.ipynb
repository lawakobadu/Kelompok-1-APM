```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
data = pd.read_csv("shopping_data.csv")
data
```

```python
cust_data = data.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(cust_data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(cust_data)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(cust_data[:,0], cust_data[:,1], c=cluster.labels_, cmap='rainbow')
plt.title('Clusters Shopping')
plt.xlabel('Annual Income')
plt.ylabel('The Spending Score')
plt.legend()
plt.show()
```
